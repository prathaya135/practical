```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('temperatures.csv')
df=df[{'ANNUAL','YEAR'}]
```

```python
df.isnull().sum()
df= pd.DataFrame(df)
x = pd.DataFrame(df['YEAR'])
y = pd.DataFrame(df['ANNUAL'])
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
result = lr.predict(X_test)
```

```python
# print('accuracy is:',lr)
print('regressor coeficient:',lr.coef_)
print('regressor intercept:',lr.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error 
mean_absolute_error(Y_test,result)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test,result)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,result)
```

```python
lr.score(X_train,Y_train)
```
